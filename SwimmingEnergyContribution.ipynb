### 3 Problema 3: diferentes estilos

Ahora es cuando se complica todo. Yo me pregunto: requiere la misma cantidad de energa nadar durante 75 segundos a crol que a braza? Y, en cualquier caso, proviene esa energa de las mismas fuentes? Bueno, la primera respuesta es no, y la segunda, no se sabe. Capelli establece el coste de nadar a los diferentes estilos mediante estas ecuaciones:

> **Mariposa:** 0.234*10^(0.547x)

> **Espalda:** 0.270*10^(0.491x)

> **Braza:** 0.548*10^(0.355x)
 
> **Crol:** 0.228*10^(0.488x)

representando la *x* el tiempo invertido en segundos.

**A continuacin puedes obtener el coste energtico en KJ/m para cualquier estilo y distancia segn el modelo de Capelli y colaboradores:**

```python
# Ajusta el tiempo y la distancia para obtener el coste energtico de desplazamiento a todos los estilos

tiempo = 75
distancia = 100

velocidad = distancia/tiempo
gasto_mariposa = round(0.234 * (10 ** (0.547 * velocidad)) ,2)
gasto_espalda = round(0.270 * (10 ** (0.491 * velocidad)) ,2)
gasto_braza = round(0.548 * (10 ** (0.355 * velocidad)) ,2)
gasto_crol = round(0.228 * (10 ** (0.488 * velocidad)) ,2)
relat_mariposa = round(((gasto_mariposa*100)/gasto_crol)-100 ,2)
relat_espalda = round(((gasto_espalda*100)/gasto_crol)-100 ,2)
relat_braza = round(((gasto_braza*100)/gasto_crol)-100 ,2)

print(f"Mariposa: {gasto_mariposa} KJ/m (+{relat_mariposa} %)")
print(f"Espalda: {gasto_espalda} KJ/m (+{relat_espalda} %)")
print(f"Braza: {gasto_braza} KJ/m (+{relat_braza} %)")
print(f"Crol: {gasto_crol} KJ/m")
```

Perfecto, le hemos puesto nmeros. Las ecuaciones de Capelli nos confirman que el gasto de energa para la misma velocidad es **muy diferente** para los distintos estilos. Todos los que hemos nadado sabemos, obviamente, que no representa la misma intensidad un 100 braza en 75 segundos que un 100 libres, precisamente por esta razn.

Tendramos ms precisin si comparramos intensidades relativas parecidas entonces? Esa es mi hiptesis. Una forma de comparar estilos en natacin, aunque con muchas comillas, son Swimming Points. Si cogemos una prueba merecedora de, por ejemplo, 900 SP a braza y otra a crol, podramos asumir que representan esfuerzos semejantes. Esto no es as, como desarrollar ms adelante, pero creo que es mejor que no hacer nada. 

De esta forma, el algoritmo completo sera:

1. Decidir el estilo y el tiempo de esfuerzo del que se quiere averiguar la contribucin relativa de los sistemas energticos.

2. Calcular su coste energtico por metro como hemos hecho en el punto anterior.

3. Averiguar sus Swimming Points.

4. Calcular la marca en la misma distancia pero a crol que es merecedora de la misma puntuacin. Esto equiparara estilos si tuvieran el mismo gasto energtico por metro.

5. Repetir el punto 2 para la marca de crol.

6. Calcular el % de diferencia en gasto energtico a crol y al otro estilo para esas marcas equivalentes.

7. Aumentar la marca de crol en ese %. Esto equiparara estilos bajo la hiptesis de que **para gastar la misma cantidad de energa a crol debes realizar un esfuerzo de mayor duracin.**

8. Obtener la contribucin de los sistemas energticos para esta marca a travs de las ecuaciones que modelan el trabajo de Rodrguez y Mader. 

**Este script representa todo ese proceso:**

```python
# Variables a modificar por el usuario
distancia = 100 # en metros
tiempo = 75 # en segundos\
estilo = 'braza' 
record_mundo_prueba_elegida = 64.13 # en segundos
record_mundo_crol = 51.71 # en segundos

# 2. Calcular el coste energtico:
velocidad = distancia/tiempo
if estilo == "mariposa":
	gasto = round(0.234 * (10 ** (0.547 * velocidad)) ,2)
elif estilo == "espalda":
	gasto = round(0.270 * (10 ** (0.491 * velocidad)) ,2)
elif estilo == "braza":
	gasto = round(0.548 * (10 ** (0.355 * velocidad)) ,2)
elif estilo == "crol":
	gasto = round(0.228 * (10 ** (0.488 * velocidad)) ,2)

# 3. Averiguar valor en Swimming Points:
SP1 = round(1000*(record_mundo_prueba_elegida/tiempo)**3, 0)

# 4. Calcular la marca en la misma distancia pero a crol que es merecedora de la misma puntuacin:
tiempo_crol = record_mundo_crol
SP2 = round(1000*(record_mundo_crol/tiempo_crol)**3, 0)

while SP2 > SP1:
	SP2 = round(1000*(record_mundo_crol/tiempo_crol)**3, 0)
	tiempo_crol += 0.01

# 5. Repetir punto 2 para la marca de crol:
velocidad_crol = distancia/tiempo_crol
gasto_crol = round(0.228 * (10 ** (0.488 * velocidad_crol)) ,2)

# 6. Calcular el % de diferencia en gasto energtico a crol y al otro estilo:
diferencia = round(((gasto*100)/gasto_crol)-100 ,2)

# 7. Aumentar la marca de crol en ese %:
tiempo_crol = tiempo_crol + tiempo_crol*(diferencia/100)

# 8. Obtener la contribucin de los sistemas energticos para esta marca:
pcr = round(424.1 * (tiempo_crol ** -0.789) + 0.8296 ,2)
glu = round(239.7 * (tiempo_crol ** -0.4676) + 1.072 ,2)
oxi = round(-585.6 * (tiempo_crol ** -0.6017) + 95.78 ,2)

print(f"Marca equivalente a crol: {round(tiempo_crol, 2)} s")
print(f"ATP-PCr: {pcr} %")
print(f"Sistema Glucolitico: {glu} %")
print(f"Sistema Oxidativo: {oxi} %")
```
